```python
MAN_PRICE=data.loc[data['neighbourhood_group']=='Manhattan'].reset_index()
BRONX_PRICE=data.loc[data['neighbourhood_group']=='Bronx'].reset_index()
BROOK_PRICE=data.loc[data['neighbourhood_group']=='Brooklyn'].reset_index()
QUEENS_PRICE=data.loc[data['neighbourhood_group']=='Queens'].reset_index()
STATEN_PRICE=data.loc[data['neighbourhood_group']=='Staten Island'].reset_index()
```

```python
MAN=MAN_PRICE.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index().sort_values('price',ascending=False)
BRON=BRONX_PRICE.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index().sort_values('price',ascending=False)
BROOK=BROOK_PRICE.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index().sort_values('price',ascending=False)
QUEENS=QUEENS_PRICE.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index().sort_values('price',ascending=False)
STATEN=STATEN_PRICE.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index().sort_values('price',ascending=False)
```

```python
man=MAN[0:3].reset_index()
bron=BRON[0:3].reset_index()
brook=BROOK[0:3].reset_index()
queens=QUEENS[0:3].reset_index()
staten=STATEN[0:3].reset_index()
```

```python
#concatenating all the indivisual datasets to form a single dataset
max_group=pd.concat([man,bron,brook,queens,staten],axis=0)
max_group
```

```python
max_group.rename(columns={'price':'mean_price'})
```

```python
from matplotlib import rcParams
rcParams['figure.figsize']=22,8
x,y = {max_group['neighbourhood'] , max_group['price']}
plt.title("price vs neighbourhood")
v=sns.barplot(x,y)
```

```python
room_avail = data.groupby(['neighbourhood_group']).agg({'availability_365': 'mean', 'price': 'mean'}).rename(columns={'availability_365': 'avg_availability_365', 'price': 'mean_price'})
room_avail
```

```python
v6=room_avail.plot(kind='bar')
plt.title('Mean v/s Room Availability')
```

FROM THE GRAPH IT IS CLEAR THAT STATEN ISLAND HAS THE HIGHEST AVAILABILITY OF ROOMS WHEREAS MANHATTAN HAS HIGHEST ROOM PRICE

```python
RD=data.groupby(['neighbourhood_group','room_type'])['room_type'].count().unstack().sort_values(by = 'Entire home/apt',ascending = False)
RD
```

```python
plt.figure(figsize=(12,6))
v2=RD.plot(kind='bar')
plt.title('Rooms distribution based on neighbourhood groups')
v2
```

  FROM THE ANALYSIS IT IS CLEAR THAT MANHATTAN HAS THE HIGHEST NUMBER OF ENTIRE HOME/APT (ROOM TYPE) WHEREAS ALL OTHER CITIES HAS HIGHEST NUMBER OF PRIVATE NUMBER OF ROOMS
